```python
import scipy.io

irfs_data = scipy.io.loadmat('Tebaldi\oo_.irfs.mat')
print(irfs_data.keys())
```

```python
# Assuming the IRFs are stored under 'oo_.irfs'
irfs = irfs_data['oo_']

# Print the shape of the IRFs array to understand its structure
print(irfs.shape)
```

```python
import numpy as np
IRF = irfs[0][0][-2]
irf_variables = {var: np.array(IRF[var][0]) for var in IRF.dtype.names}
# Now, each variable is stored in `irf_variables` dictionary

Y_e = irf_variables['Y_e'][0][0]
C_e = irf_variables['C_e'][0][0]
A_e = irf_variables['A_e'][0][0]
LA_e = irf_variables['LA_e'][0][0]
PA_e = irf_variables['PA_e'][0][0]
K_e = irf_variables['K_e'][0][0]
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 2, figsize=(12, 12), sharex = True)
# Lista de variables y nombres
variables = [Y_e, A_e, LA_e,C_e, PA_e, K_e]
titles = ['a. Producción Bien Final', 'b. Consumo Hogar', 'c. Capital Privado',
          'd. Capital Humano en I+D', 'e. Precio Blueprints', 'f. Tecnología']

# Plotear cada variable
for i, ax in enumerate(axes.flat):
    ax.plot(variables[i], label=titles[i], color='r', linewidth=2)
    ax.axhline(0, color='gray', linestyle='--', linewidth=1)
    ax.set_title(titles[i], fontsize=12, fontname='Arial', loc='left')
    ax.set_xlabel("Periodo", fontsize=12, fontname='Arial')
    ax.set_ylabel("Respuesta", fontsize=12, fontname='Arial')
    ax.grid(ls ='dashed',lw=0.5)
    ax.set_xlim(-.5,150)

# Ajustar el diseño
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('Imagenes/IRF.png', dpi=300, bbox_inches='tight')

plt.show()
```
