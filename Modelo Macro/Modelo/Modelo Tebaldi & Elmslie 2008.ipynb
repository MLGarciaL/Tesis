### 4.2.1 Sector Bien Final

Siguiendo el planteamiento de los autores mencionados previamente, la firma representativa produce un bien final utilizando una tecnología de rendimientos constantes a escala y operando en un mercado de competencia perfecta. La elección de una tecnología de rendimientos constantes asegura que el valor del producto total sea proporcional a la cantidad de insumos empleados, lo cual también implica competitividad en el mercado.

$$Y = L^{1-\alpha} \int_0^{f(A,T)} x_i^\alpha di\qquad\qquad\qquad\text{(1)}$$

donde $L$ es la cantidad de trabajo empleado en el sector de bien final, $x_i$ denota la cantidad de productos intermedios utilizados en la producción, $A$ representa el nivel de tecnología disponible, mientras que $T$ refleja la calidad institucional. Además, la forma de la función de producción adopta el enfoque de Dixit y Stiglitz (1977), al incorporar un sector de productos intermedios caracterizado por competencia monopolística. Este tipo de tecnología indica que el progreso tecnológico se encuentra en la diversificación de productos. El hecho de que la función de producción sea aditivamente separable implica que cada nuevo bien intermedio generado se considera como un insumo adicional que amplía la gama de productos, pero no mejora ni sustituye a los existentes. En otras palabras, los nuevos bienes de capital son distintos de los anteriores, pero no superiores ni inferiores, por lo que los productos antiguos mantienen su relevancia y nunca quedan obsoletos, Sala I Martin (2000).

La razón de integrar hasta $f(A,T)$ sigue la misma lógica que en el trabajo original de Tebaldi y Elmslie (2006). Esto implica que los nuevos productos intermedios no solo dependen de la cantidad de innovaciones generadas ($A$), sino también de la evolución conjunta de las instituciones ($T$), que posibilitan la adopción y difusión efectiva de estas innovaciones en el sistema productivo. 

Siguiendo el mismo razonamiento de Tebaldi y Elmslie, la ecuación $(1)$ se puede reescribir como:

$$Y = L_Y^{1-\alpha} \int_0^{\min\{A,\psi T\}} x_i^\alpha di = L_Y^{1-\alpha} \int_0^{\psi T} x_i^\alpha di\qquad\qquad\qquad\text{(2)}$$

Esto implica que la tecnología avanza a un ritmo más rápido que las instituciones, lo que refleja una dinámica característica de las economías modernas donde la innovación tecnológica supera la capacidad de las instituciones para adaptarse. Si, en cambio, las instituciones evolucionaran más rápidamente que la tecnología, el modelo podría reducirse a una versión del modelo de Romer (1990), en el que el progreso técnico es exógeno y las instituciones no jugarían un papel determinante en el ritmo de adopción y difusión de las innovaciones.

Al asumir un mercado perfectamente competitivo y un comportamiento maximizador por parte de las firmas, estas buscan maximizar sus beneficios, definidos como la diferencia entre los ingresos totales y los costos de los insumos utilizados en la producción. Lo que nos permite obtener las siguientes condiciones del modelo

$\max \pi_Y = L^{1-\alpha} \int_0^{\psi T} x_i^\alpha di - wL - \int_0^{\psi T} p_ix_i di$

**CPO**

$$w = \left(1-\alpha\right) \frac{Y}{L_Y} \qquad\qquad\qquad\text{(3)}$$

$$p_i = \alpha L_Y^{1-\alpha} x_i^{\alpha-1}\qquad\qquad\qquad\text{(4)}$$

La ecuación $(3)$ y $(4)$ corresponden a la inversa de la demanda de trabajador y de bienes intermedios, correspondientemente.

### 4.2.2 Sector Bien Intermedio


El sector de bienes intermedios está compuesto por un gran número de firmas, cada una de las cuales opera como un monopolista en la producción de un bien intermedio específico. Este supuesto sigue el planteamiento clásico de Romer (1990), donde los bienes intermedios se diferencian entre sí y son producidos utilizando una tecnología exclusiva, lo que otorga poder de mercado a las firmas productoras. Así, cada una de ellas enfrenta a una demanda propia, que está caracterizada por la ecuación $(4)$. 

Además, el sector de bienes intermedios utiliza exclusivamente capital para la producción, de tal manera que, una vez adquirida la patente, cada unidad de $x_i$ se produce utilizando una unidad de capital $K$. Por lo tanto, el costo que enfrenta el productor corresponde al interés que debe pagar por el capital prestado. Bajo el supuesto de comportamiento maximizador, las siguientes condiciones de equilibrio se derivan:

$\max \pi_Y = p_i(x_i)x_i - rx_i = \alpha L_Y^{1-\alpha} x_i^{\alpha} - rx_i $


**CPO**

$\alpha^2 L_Y^{1-\alpha} x_i^{\alpha-1} = r$

$$x_i =\left(\frac{\alpha^2 L_Y^{1-\alpha}}{r}\right)^{1/(1-\alpha)}\qquad\forall i\qquad\qquad\text{(5)}$$

Sustituyendo $(5)$ en $(4)$ obtenemos:

$p_i = \alpha L_Y^{1-\alpha}\left(\frac{\alpha^2 L_Y^{1-\alpha}}{r}\right)^{(\alpha-1)/(1-\alpha)}$

$$p_i = \frac{r}{\alpha}\qquad\forall i\qquad\qquad\text{(4')}$$

Además, el costo de la patente que la firma intermedia debe adquirir para poder producir está determinado por el valor presente descontado de los beneficios futuros que generará. 

$P_A = \sum_{i=\tau}^{\infty} \frac{\pi_Y}{(1+r)^i}=\frac{\pi_Y}{(1+r)^\tau}\sum_{i=0}^{\infty} \frac{1}{(1+r)^i} $

$\pi_i = rx_i\left(\frac{1}{\alpha} -1\right) = \frac{1-\alpha}{\alpha}rx_i$

$$P_A = \frac{1}{(1+r)^{\tau-1}}\frac{1-\alpha}{\alpha}x_i \qquad\qquad\qquad\text{(6)}$$


El parámetro $\tau$ indica el tiempo que debe transcurrir antes de que una nueva innovación pueda ser incorporada por las firmas intermedias. Si $\tau = 0$, las firmas pueden utilizar la innovación de forma inmediata tras su adquisición, lo que sugiere un entorno institucional favorable, sin barreras temporales significativas.

### 4.2.3 Sector Investigación y Desarrollo

El sector de investigación y desarrollo (I+D) es responsable de generar innovación mediante la creación de nuevos productos intermedios y avances tecnológicos. Ellos venden las patentes a las firmas de bienes intermedios, y utilizan trabajadores para su producción de esta forma nos queda que el producto innovación es igual a:

$$A_{t+1} - A_{t} = \delta_A A_t^{1-\beta}\left(\psi  T_t\right)^\beta {L_A}_t \qquad\qquad\qquad\text{(7)}$$

Las firmas maximizan el beneficio de su producto, por lo tanto podemos obtener la siguiente condición de orden:

$\max \pi_A = P_A\left(\delta_A A^{1-\beta}\left(\psi  T_t\right)^\beta {L_A}\right) - wL_A $

$$P_A\delta_A A^{1-\beta}\left(\psi  T\right)^\beta = w\qquad\qquad\qquad\text{(8)}$$

### 4.2.4 Consumidores

Finalmente, este sector, al igual que los anteriores, se modela asumiendo un comportamiento maximizador, en este caso respecto a la utilidad del consumidor. Para representar dichas preferencias, se emplea una función de utilidad CES (Constant Elasticity of Substitution), dada su flexibilidad para capturar distintas elasticidades de sustitución, tanto entre bienes como a lo largo del tiempo. Además, su adopción es ampliamente aceptada en la literatura económica, lo que favorece una interpretación consistente del modelo y facilita su comparación con enfoques similares.

$$U_t  = \gamma \frac{C_t^{1-\eta_C} - 1}{1-\eta_C} + \theta \frac{(1-L_t)^{1-\eta_L} - 1}{1-\eta_L} \qquad\qquad\qquad\text{(9a)}$$

Observemos que según el cuadro las familias se enfrentan a la siguiente restricción intertemporal

$$C_t + I_t = w_t\left({L_A}_t + {L_Y}_t\right) + r_tK_t + \Pi_t \qquad\qquad\qquad\text{(9b)}$$

Donde es claro que:

$${L_A}_t + {L_Y}_t = L_t\qquad\qquad\qquad\text{(10)}$$

Asimismo, los consumidores dueños del capital presentan la siguiente ecuación  de movimiento:

$$K_{t+1} - K_t = I_t - \delta_K K_t\qquad\qquad\qquad\text{(11)}$$


Suponiendo que las familias viven infinitos periodos, y descuentan intertemporalmente a la tasa $\rho$, tal que $\beta \equiv \frac{1}{1+\rho}$, utilizando las ecuaciones $\text{(9a)}$ y $\text{(9b)}$,  nos quedan las siguientes condiciones de equilibrio:

$\max \sum_{t=0}^\infty  \beta^t\left(\gamma \frac{C_t^{1-\eta_C} - 1}{1-\eta_C} + \theta \frac{(1-L_t)^{1-\eta_L} - 1}{1-\eta_L}\right)~~\text{ s.t.} C_t + K_{t+1} - \left(1-\delta_K\right)K_t = w_tL_t + r_tK_t + \Pi_t$

**CPO**

$\frac{\partial \mathscr L}{\partial C_t} = \beta^t\gamma C_t^{-\eta_C} -\lambda_t = 0$

$\frac{\partial \mathscr L}{\partial L_t} = -\beta^t\theta \left(1-L_t\right)^{-\eta_L} + w_t\lambda_t = 0$

$\frac{\partial \mathscr L}{\partial K_{t+1}} = - \lambda_t  + \lambda_{t+1}\left(1+r_{t+1} - \delta_K\right)= 0$

$$C_{t+1} = \left[\beta{\left(1+r_{t+1} - \delta_K\right)}\right]^{1/\eta_C} C_{t}  \qquad\qquad\qquad\text{(12)}$$

$$w_{t} = \frac{\theta}{\gamma} \frac{C_t^{\eta_C}}{\left(1-L_t\right)^{\eta_L} } \qquad\qquad\qquad\text{(13)}$$

Donde $\text{(12)}$ es la ecuación de Euler y  $\text{(13)}$ la oferta de trabajo.

Teniendo todas las partes del sistema podemos proceder al cerrado del modelo.

### 4.2.5 Vaciado de Mercado

Dada la ecuación $\text{(4')}$, tenemos que el precio de todos los bienes intermedios son igual a una misma constante $p = p_i= \frac{r}{a}~\forall i$, y lo mismo nos queda para $\text{(5)}$, $x_i = x$. Integrando $x_i$ con respecto a todos los bienes intermedios utilizados en la economía y definiendo $z = \frac{\psi T}{A}$, nos queda: 

$\int_0^{\psi T}x_i di = K = \psi T \left(\frac{\alpha^2}{r}\right)^{1/(1-\alpha)} L_Y $

$$\int_0^{\psi T}x_i di = K \implies K = zAx \qquad\qquad\qquad\text{(14)}$$

Utilizando $\text{(14)}$  podemos escribir a la ecuación $\text{(2)}$ de la siguiente forma:

$Y =  \psi T L_Y^{1-\alpha}  x^\alpha $

$$Y =  zA L_Y^{1-\alpha}  x^\alpha \qquad\qquad\qquad\text{(2')}$$

Además, podemos reescribir a las ecuaciones de la siguiente forma:

$$A_{t+1} - A_{t} = \delta_AA_tz^\beta {L_A}_t \qquad\qquad\qquad\text{(7')}$$

$$P_A\delta_A A z^\beta = w\qquad\qquad\qquad\text{(8')}$$

De esta forma nos queda que las ecuaciones $\text{(2)}$, $\text{(3)}$, $\text{(4')}$, $\text{(5)}$, $\text{(6)}$, $\text{(7')}$, $\text{(8')}$, $\text{(9b)}$, $\text{(10)}$, $\text{(11)}$, $\text{(12)}$ y $\text{(13)}$, nos deja un sistema de 12 ecuaciones y 12 incognitas.
