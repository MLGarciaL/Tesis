# Innovation, growth and quality of life: a theoretical model and an estimate for the Italian regions
##  d'Agostino & Scarlato

### Modelo Ramsey:


$$\max_{c_t} \sum_{t=0}^\infty \beta^t\frac{c_t^{1-\sigma}}{1-\sigma}$$

Función de producción bien de consumo

$$Y = F(L_Y,x_i) =(AL_Y)^{\alpha}\int_0^A x_i^{1-\alpha} di$$

Maximización del beneficio:

$$\max \pi_Y = (AL_Y)^{\alpha}\int_0^A x_i^{1-\alpha} di - wL_Y - p_ix_i$$

**CPO**

$\frac{\partial \pi_Y}{\partial L_Y} = \alpha A\left((AL_Y)^{\alpha}\int_0^A x_i^{1-\alpha} di\right) L_Y^{-1} - w = 0$

$\frac{\partial \pi_Y}{\partial x_i} = \left(1-\alpha\right) (AL_Y)^{\alpha}x_i^{-\alpha} - p = 0 \qquad\qquad \forall i$

Por lo tanto, las condiciones de equilibrio son:

$$w = \alpha \frac{Y}{L_Y}$$

$$p_i = \left(1-\alpha\right) \left(\frac{AL_Y}{x_i}\right)^\alpha\forall i$$

Beneficio bien intermedio:

$$\max_{x_i}\pi_{i} = p(x_i)_i x_i - rx_i$$

**CPO**

$\frac{\partial \pi_i}{\partial x_i} = \left(1-\alpha\right)^2 \left(\frac{AL_Y}{x_i}\right)^\alpha - r =0 $

$x_i^\alpha =  \left(1-\alpha\right)^2 \frac{\left(AL_Y\right)^\alpha}{r}$

$$x_i = \left(\frac{\left(1-\alpha\right)^2}{r}\right)^\frac{1}{\alpha} AL_Y$$

Reemplazando en la  condición de $p_i$ nos queda:

$$\hat p_i = \frac{r}{1-\alpha}\forall i$$

Por lo tanto:

$$\hat x_i = \hat x = \left[\frac{(1-\alpha)}{\hat p}\right]^{1/\alpha}AL_Y\quad\quad \forall i$$

Obteniendo la función de beneficio óptimo:

$$\hat \pi_i =r\hat x_i\left(\frac{1}{1-\alpha} - 1\right) = r\hat x_i \frac{\alpha}{1-\alpha}$$

### Modelo Ramsey:

$$Y = F(L_Y,x_i) =(AL_Y)^{\alpha}\left(\sum_0^A x_i^{1-\alpha}\right)$$

Maximización del beneficio:

$$\max \pi_Y = (AL_Y)^{\alpha}\left(\sum_0^A x_i^{1-\alpha}\right) - wL_Y - p_ix_i$$

$\frac{\partial \pi_Y}{\partial L_Y} = \alpha \left[(AL_Y)^{\alpha}\left(\sum_0^A x_i^{1-\alpha}\right)\right] L_Y^{-1} - w = 0$

$\frac{\partial \pi_Y}{\partial x_i} = \left(1-\alpha\right) (AL_Y)^{\alpha}x_i^{-\alpha} - p = 0 \qquad\qquad \forall i$

Por lo tanto, las condiciones de equilibrio son:

$$w = \alpha \frac{Y}{L_Y}$$

Beneficio bien intermedio:

$$\max_{x_i}\pi_{i} = p(x_i)_i x_i - rx_i$$

$\frac{\partial \pi_i}{\partial x_i} = \left(1-\alpha\right)^2 \left(\frac{AL_Y}{x_i}\right)^\alpha - r =0 $

$x_i^\alpha =  \left(1-\alpha\right)^2 \frac{\left(AL_Y\right)^\alpha}{r}$

$$x_i = \left(\frac{\left(1-\alpha\right)^2}{r}\right)^\frac{1}{\alpha} AL_Y$$
