```python
df_free = pd.read_excel('Bases\Freedom.xlsx')
df_free = df_free[['Year','ISO_Code','Area 2']]
df_free = df_free.rename(columns = {'ISO_Code':'Code','Countries':'Entity','Area 2':'Riesgo de Expropiación'})
df_free.head()
```

```python
df = df_inst.merge(df_free, on= ['Year','Code'], how = 'left')
df
```
