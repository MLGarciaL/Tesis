```python
import pickle

with open('Bases\- Simulaci√≥n n = 3ER.pkl', 'rb') as f:
    datos = pickle.load(f)
```

```python
At = datos['At']
Atmax = datos['Atmax']
patentes_totales = datos['patentes_totales']
At1 = datos['At1']
Atmax1 = datos['Atmax1']
patentes_totales1 = datos['patentes_totales1']
At2 = datos['At2']
Atmax2 = datos['Atmax2']
patentes_totales2 = datos['patentes_totales2']
At3 = datos['At3']
Atmax3 = datos['Atmax3']
patentes_totales3 = datos['patentes_totales3']
At4 = datos['At4']
Atmax4 = datos['Atmax4']
patentes_totales4 = datos['patentes_totales4']
At5 = datos['At5']
Atmax5 = datos['Atmax5']
patentes_totales5 = datos['patentes_totales5']

```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial'
import numpy as np
import matplotlib.ticker as ticker


fig, axes = plt.subplots(1, 3, figsize=(10.5, 3.5), sharex=True, sharey=True)

#fig.suptitle(r'Distribuci√≥n de Patentes Totales $n=2$', fontsize=14, fontname='Arial')

axes[0].hist(patentes_totales2, bins=20, color='white', edgecolor='black')
axes[0].axvline(np.mean(patentes_totales), color='red', linestyle='dashed', linewidth=2, alpha=0.2)
axes[0].axvline(np.mean(patentes_totales1), color='red', linestyle='dashed', linewidth=2, alpha=0.2)
axes[0].axvline(np.mean(patentes_totales4), color='blue', linestyle='dashed', linewidth=2, alpha=0.2)
axes[0].axvline(np.mean(patentes_totales3), color='blue', linestyle='dashed', linewidth=2, alpha=0.2)
axes[0].axvline(np.mean(patentes_totales2), color='red', linestyle='dashed', linewidth=2.5, label = r'$pat = 0, I = 1 $', zorder=2)
axes[0].axvline(np.mean(patentes_totales5), color='blue', linestyle='dashed', linewidth=2.5, label = r'$pat = 0, I = 0 $', zorder=1)
axes[0].set_ylabel('Frecuencia', fontsize=12, fontname='Arial')
axes[0].set_title('a')
axes[0].legend(fontsize=8)

axes[1].hist(patentes_totales1, bins=20, color='white', edgecolor='black')
axes[1].axvline(np.mean(patentes_totales), color='red', linestyle='dashed', linewidth=2, alpha=0.2)
axes[1].axvline(np.mean(patentes_totales2), color='red', linestyle='dashed', linewidth=2, alpha=0.2)
axes[1].axvline(np.mean(patentes_totales3), color='blue', linestyle='dashed', linewidth=2, alpha=0.2)
axes[1].axvline(np.mean(patentes_totales5), color='blue', linestyle='dashed', linewidth=2, alpha=0.2)
axes[1].axvline(np.mean(patentes_totales1), color='red', linestyle='dashed', linewidth=2.5, label = r'$pat = 5, I = 1 $', zorder=2)
axes[1].axvline(np.mean(patentes_totales4), color='blue', linestyle='dashed', linewidth=2.5, label = r'$pat = 5, I = 0 $', zorder=1)
axes[1].set_xlabel('Patentes', fontsize=12, fontname='Arial')
axes[1].set_title('b')
axes[1].legend(fontsize=8)

axes[2].hist(patentes_totales, bins=20, color='white', edgecolor='black')
axes[2].axvline(np.mean(patentes_totales1), color='red', linestyle='dashed', linewidth=2, alpha=0.2)
axes[2].axvline(np.mean(patentes_totales2), color='red', linestyle='dashed', linewidth=2, alpha=0.2)
axes[2].axvline(np.mean(patentes_totales4), color='blue', linestyle='dashed', linewidth=2, alpha=0.2)
axes[2].axvline(np.mean(patentes_totales5), color='blue', linestyle='dashed', linewidth=2, alpha=0.2)
axes[2].axvline(np.mean(patentes_totales), color='red', linestyle='dashed', linewidth=2.5, label = r'$pat = 10, I = 1 $', zorder=2)
axes[2].axvline(np.mean(patentes_totales3), color='blue', linestyle='dashed', linewidth=2.5, label = r'$pat = 10, I = 0 $', zorder=1)
axes[2].set_title('c')
axes[2].legend(fontsize=8)
axes[0].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:,.1f}".replace(".", ",")))    
axes[1].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:,.1f}".replace(".", ",")))    
axes[2].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:,.1f}".replace(".", ",")))    


plt.savefig('Imagenes/distribucion_patentesn=3ER.png', dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import mannwhitneyu

stat, p_value = mannwhitneyu(patentes_totales, patentes_totales1, alternative='greater')
print('TP =10, I = 1 ‚ÄîTP= 5,I=1')
print(stat, round(p_value,3))
print('')

stat, p_value = mannwhitneyu(patentes_totales1, patentes_totales2, alternative='greater')
print('TP = 5,I=1‚ÄîTP= 0,I=1')
print(stat, round(p_value,3))
print('')

stat, p_value = mannwhitneyu(patentes_totales, patentes_totales3, alternative='greater')
print('TP =10, I = 1 ‚ÄîTP=10,I=0')
print(stat, round(p_value,3))
print('')

stat, p_value = mannwhitneyu(patentes_totales1, patentes_totales4, alternative='greater')
print('TP = 5,I=1‚ÄîTP= 5,I=0')
print(stat, round(p_value,3))
print('')

stat, p_value = mannwhitneyu(patentes_totales2, patentes_totales5, alternative='greater')
print('TP = 0,I=1‚ÄîTP= 0,I=0')
print(stat, round(p_value,3))
print('')
```

```python
import numpy as np
from scipy.stats import ttest_ind, levene, mannwhitneyu

# Suponiendo que At5 y At son listas de listas de dimensiones [200 x 102]
# Iterar sobre el tiempo (dimensi√≥n 2)
for t in [5,50, 100]:  # t = 0, 1, ..., 101 (102 tiempos)
    print(f"\nTiempo {t}:")

    # Extraer los valores correspondientes al tiempo t de todas las listas
    At5_t = [lista[t] for lista in Atmax5]  # Valores en At5 en tiempo t
    At_t = [lista[t] for lista in Atmax]    # Valores en At en tiempo t

    # Prueba de Levene
    levene_stat, p_value_levene = levene(At5_t, At_t)

    # Prueba de Mann-Whitney
    u_stat, p_value_mannwhitney = mannwhitneyu(At5_t, At_t)

    # Mostrar resultados
    print(f"  Prueba de Levene: Estad√≠stico = {levene_stat:.0f}, p-valor = {p_value_levene:.3f}")
    print(f"  Prueba Mann-Whitney: Estad√≠stico U = {u_stat}, p-valor = {p_value_mannwhitney:.3f}")
```
