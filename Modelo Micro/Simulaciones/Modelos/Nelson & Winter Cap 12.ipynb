## Oferta Individual

$$Q_{it} = A_{it}K_{it}$$

El capital es el único factor de producción, y tiene un costo constante $c$ hay retornos constantes a escala.

```python
def quantity_it(A_it,K_it):
    return A_it*K_it
```

Donde la cantidad total queda determinada por:

$$Q_t = \sum Q_{it} = \sum A_{it}K_{it}$$


```python
def price_t(Q_t, D, eta):
    return D/Q_t**(1/eta)
```

El beneficio por unidad de capital de una firma es igual a:

$$\pi_{it} = \left(P_tA_{it} - c - r_{im} - r_{in}\right)$$

que implica los ingresos menos los costos, menos los recursos utilizados en imitación e innovación.

Los beneficios por lo tanto:

$$\Pi_{it} = \pi_{it}\cdot K_{it}$$

```python
def profit_rate(A_it, p_t, r_im, r_in, c):
    return p_t*A_it - c - r_im - r_in
    
def profit(pi_it, K_it):
    return pi_it*K_it
```

Las actividades de investigación y desarrollo generan nuevos niveles de productividad mediante un proceso aleatorio de dos etapas. La primera etapa esta caracterizada por variables independientes aleatorias $d_{imt}$ y $d_{int}$, que toman un valor de uno o cero de acuerdo si la firma $i$ decide o no imitar o innovar en el periodo $t$. El exíto de obtener tales resultados ocurren con las respectivas probabilidades:

$$Pr(d_{imt} = 1) = a_m r_{im}K_{it}$$

$$Pr(d_{int} = 1) = a_n r_{in}K_{it}$$

```python
def dummy_imt(r_im, K_it, a_m):
    prob_imt = a_m*r_im*K_it
    return np.random.choice([0, 1], p=[1 - prob_imt, prob_imt])
    
def dummy_int(r_in, K_it, a_n):
    prob_int = a_n*r_in*K_it
    return np.random.choice([0, 1], p=[1 - prob_int, prob_int])
```

Si una empresa obtiene un atractivo para la innovación, toma muestras de una distribución de oportunidades tecnológicas, $F(A;t, A_{it})$. Esta distribución es  una función del tiempo y es independiente de la técnica de la firma actual. 

$$\tilde A_{it} \sim F(A;t,A_{it})$$

Por lo tanto, debido a las innovaciones basadas en la ciencia:

$$\log \left(\tilde A_{it}\right) \sim N\left(\lambda(t), \sigma^2\right)$$


$\lambda(t) = 0.16 + 0.01t$

$\sigma = 0.05$

El tipo de innovación es independiente de la cantidad de recursos implementados. Y solo depende del tiempo y de la tecnología que tiene la empresa.

```python
def innov(t):
    lambda_t = 0.16 + 0.01*t
    sigma = np.sqrt(0.5)
    return np.random.normal(loc=lambda_t, scale=sigma)
```

```python
import matplotlib.pyplot as plt

t_values = np.arange(0, 100, 10)  # Valores de t de 0 a 50 con paso 5
samples = {t: [innov(t) for _ in range(1000)] for t in t_values}  # 1000 muestras por cada t

# Graficar la distribución de las muestras para cada t
plt.figure(figsize=(12, 8))
for t, data in samples.items():
    plt.scatter([t] * len(data), np.log(data), alpha=0.1)

plt.plot([0,90],[np.mean(samples[0]),np.mean(samples[90])],c='red')
plt.ylabel("Valor de innov(t)")
plt.xlabel("x")
plt.grid(True)
plt.show()
```

Si una firma obtiene imitación, entonces tiene la opción de observar y copiar la mejor práctica de la industria. De esta forma podemos definir de la siguiente forma:

$$\tilde A_{it} = A_{i,t} + d_{int}\cdot \left(e^{\log \left(\tilde A_{it}\right)} - A_{i,t}\right)$$

$$\hat A_{it} = A_{i,t} + d_{imt}\cdot \left(A_t^* - A_{i,t}\right)$$



Para las firmas que obtienen ambos imitación e innovación en un periodo particular, el nivel de productividad del periodo siguiente esta dado por:

$$A_{i,t+1} = \max\{A_{it},\hat A_t, \tilde A_{it}\}$$

Donde $\hat A_t$ es la mejor práctina en la industria en el periodo $t$, y $\tilde A_{it}$ es una variable aleatoria como resultado de la gannancia de innovación. Por supuesto, la empresa puede no lograr un atractivo imitativo, un atractivo innovador, o ambos, en cuyo caso el menú del cual se extrae la productividad del siguiente período es más corto.

La decisión de una firma de expandirse o contraerse esta determinada por la proporción precio-producto, $P/(c/A)$, o equivalentemente el porcentaje margen sobre costo y su proporción de mercado. 

```python
def A_it1(A_it, d_int, d_imt, t, Amax):
    logtildeA_it = innov(t)
    tildeA_it = A_it + d_int*(logtildeA_it - A_it)
    
    hatA_it = A_it + d_imt*(Amax - A_it)
    return max(A_it, hatA_it, tildeA_it)
```

### Función de Inversión

$$K_{i,t+1} = I\left(\frac{P_tA_{i,t+1}}{c}, \frac{Q_{it}}{Q_t}, \pi_{it},\delta\right)\cdot K_{it} + \left(1-\delta\right)K_{it}$$


$\rho = \frac{p_tA_{i,t+1}}{c}$

$s \equiv \frac{Q_{it}}{Q_t}$

$$I = \max \left\{0,\min\{1.03 - \frac{2-s}{\rho(2-2s)}\}, f(\pi) \right\}$$

(Asumiendo que no hay intervención del banco.

$$f(\pi) =\left|\begin{align} &  \delta + \pi & \pi\leq 0\\ & \delta + 2\pi & \pi> 0\\ & \delta + 3.5\pi & \pi> 0\end{align}\right.$$

```python
def investment(p_t, A_it, Q_t, q_it, pi_it, delta, c, BANK = 0):
    if pi_it<=0:
        fpi = delta + pi_it
    else:
        fpi = delta + (1 + BANK)*pi_it
    rho = p_t*A_it/c
    s = q_it/Q_t
    return max(0,min(1 + delta - (2-s)/(rho*(2-2*s)),fpi))
```

```python
def capitaldynamic(K_it, I_it, delta):
    return (1 - delta + I_it)*K_it
```

```python
def NWCap12(t = 100, n = 32, prop = 0.5, eta = 1, D = 67, c = .16, a_m = 1.25, a_n = .125, delta = 0.03, K0 = 139.58, A0 = 0.16):
    nint = round(n*prop)

    dimt_it = np.zeros((n, t + 1))
    dint_it = np.zeros((n, t + 1))
    Pi_it = np.zeros((n, t + 1))
    pi_it = np.zeros((n, t + 1))
    A_it = np.zeros((n, t + 2))
    K_it = np.zeros((n, t + 2))
    q_it = np.zeros((n, t + 1))
    I_it = np.zeros((n, t + 1))
    p_t = np.zeros(t + 1)
    Q_t = np.zeros(t + 1)
    
    
    r_im = 0.00097
    r_in = 0.0194
    
    K_it[:, 0] = K0
    A_it[:, 0] = A0
    
    for t in range(0, t + 1):
        Amax = max(A_it[:, t])

        # Firmas Innovadoras-Imitadoras
        for i in range(0,nint):
            q_it[i,t] = quantity_it(A_it[i,t],K_it[i,t])
            dimt_it[i,t] = dummy_imt(r_im, K_it[i,t], a_m)
            dint_it[i,t] = dummy_int(r_in, K_it[i,t], a_n)
            A_it[i,t+1] = A_it1(A_it[i,t], dimt_it[i,t], dint_it[i,t], t, Amax)

        # Firmas Imitadoras
        for i in range(nint,n):
            q_it[i,t] = quantity_it(A_it[i,t],K_it[i,t])
            dimt_it[i,t] = dummy_imt(r_im,K_it[i,t], a_m)
            dint_it[i,t] = 0
            A_it[i,t+1] = A_it1(A_it[i,t], dimt_it[i,t], dint_it[i,t], t, Amax)
        
        Q_t[t] = sum(q_it[:, t])
        p_t[t] = price_t(Q_t[t], D, eta)
        
        for i in range(0,n):
            pi_it[i,t] = profit_rate(A_it[i,t], p_t[t], r_im, r_in, c)
            Pi_it[i,t] = profit(pi_it[i,t], K_it[i,t])
            I_it[i,t] = investment(p_t[t], A_it[i,t+1], Q_t[t], q_it[i,t], pi_it[i,t], delta, c)
            K_it[i,t+1] = capitaldynamic(K_it[i,t], I_it[i,t], delta)

    return NWCap12Results(dimt_it, dint_it, Pi_it, pi_it, A_it, K_it, q_it, I_it, p_t, Q_t)
```

```python
class NWCap12Results:
    def __init__(self, dimt_it, dint_it, Pi_it, pi_it, A_it, K_it, q_it, I_it, p_t, Q_t):
        self.dimt_it = dimt_it
        self.dint_it = dint_it
        self.Pi_it = Pi_it
        self.pi_it = pi_it
        self.A_it = A_it
        self.K_it = K_it
        self.q_it = q_it
        self.I_it = I_it
        self.p_t = p_t
        self.Q_t = Q_t
```

```python
resultados = NWCap12()
resultados2 = NWCap12(a_m = 4)
```

```python
Aprom_t = np.mean(resultados.A_it, axis=0)
Amax_t = np.max(resultados.A_it, axis=0)
Amin_t = np.min(resultados.A_it, axis=0)

Aprom_t2 = np.mean(resultados2.A_it, axis=0)
Amax_t2 = np.max(resultados2.A_it, axis=0)
Amin_t2 = np.min(resultados2.A_it, axis=0)
```
