El modelo se enfoca en describir el comportamiento de las firmas en una industria donde todas enfrentan una misma demanda, con una dinámica similar a la de un duopolio de Cournot. En este contexto, las firmas producen un producto homogéneo, diferenciándose únicamente por las técnicas de producción empleadas. Se asume que el capital es el único factor de producción, aunque esta simplificación puede extenderse al incorporar otros factores que dependen de manera proporcional al capital, manteniendo así la coherencia del análisis.

$$q_{it} = A_{it}\cdot K_{it}\quad\forall i\tag{1}$$

Asimismo, las empresas se enfrentan a una misma demanda con pendiente negativa, aunque esta se torna completamente elástica al superar el precio $P_{\max}$. De esta forma la inversa de la demanda toma la forma de:

$$P_t = \min\left\{\frac{D}{Q_t^\frac{1}{\eta}}, P_{\max}\right\}\tag{2}$$

Donde $Q_t = \sum_{i=1}^n q_{it}$, es decir, representa la producción total de la industria como la suma de las producciones individuales de las firmas. 
Además, las empresas asignan recursos a actividades de investigación y desarrollo (I+D). En este marco, las firmas innovadoras tienen la capacidad de generar nuevas tecnologías y adoptarlas en su producción, siempre que estas sean superiores a las empleadas previamente. Asimismo, las empresas pueden observar las tecnologías funcionales desarrolladas por otras firmas e invertir recursos en intentar imitar aquellas disponibles luego de expirar su patente. Por otro lado, las firmas imitadoras, solo destinan recursos a I+D para copiar las tecnologías ya desarrolladas por otras empresas. 

Por lo tanto, la empresa enfrenta costos de capital $c$, los cuales son fijos por unidad de capital, así como $r_{int}$ y $r_{imt}$, que representan la cantidad de recursos por unidad de capital asignados a las actividades de investigación y desarrollo para la innovación e imitación de tecnologías, respectivamente. Es claro, que las empresas dedicadas únicamente a imitar, no dedican recursos a innovar, por lo tanto, $r_{int} = 0$. A diferencia del modelo original, se introduce además un componente adicional: costos de transacción asociados a las operaciones de la firma por unidad de capital, determinados por los niveles institucionales, representados por $\xi_{TC}$. 

La función de beneficios por unidad de capital de la firma queda definida como: 

$$\pi_{it} = P_t \cdot K_{it} - c - r_{int} - r_{imt} + \xi_{TC} \cdot INS$$
 

Donde $INS$ es una variable dicotómica que toma el valor de 1 en presencia de instituciones y 0 en su ausencia. El término $\xi_{TC} \cdot INS$ aparece con un signo positivo en la función de beneficios porque refleja los efectos favorables que las instituciones pueden tener en la actividad económica de las firmas. En presencia de instituciones $INS = 1$, se asume que las empresas enfrentan menores costos de transacción gracias a un entorno institucional más eficiente y estable. La justificación de esta variable se fundamenta particularmente en los trabajos de Coase (1937) y North (1990)...

Al final de cada periodo, las firmas evalúan su desempeño. $X_{it}$, lo que determinará su decisión de continuar en la industria, ajustar sus políticas de investigación y desarrollo, o mantener su presencia en el mercado sin modificaciones. Este proceso de evaluación les permite adaptar sus estrategias en función de los resultados obtenidos y las condiciones del entorno.

$$X_{it} = \theta X_{i(t-1)} + (1-\theta)\pi_{it}$$

El cual es un promedio ponderado del desempeño en el periodo pasado y las ganancias del periodo actual. 

El proceso de cambio tecnológico esta determinado por un "proceso aleatorio de dos etapas", como lo denomina Winter (1983). En una primera etapa se determina si los procesos de investigación, innovación e imitación, fueron exitosos, luego se determina que nivel de tecnología corresponde al proceso de innovación. La posibilidad de que la actividad sea exitosa depende tanto de la cantidad total de recursos destinados, es decir, de la política y de la cantidad de capital de la empresa:

$$Pr[d_{imt} = 1] = a_m\cdot r_{imt}\cdot K_{imt}$$

$$Pr[d_{int} = 1] = a_n\cdot r_{int}\cdot K_{int}$$

Una vez que, $Pr[d_{int} = 1]$, la segunda etapa del modelo determina la tecnología descubierta, que se toma como una muestra de una distribución lognormal definida por los siguientes parámetros:
$$\log A_{i,t} \sim N\left[L\left(1+\xi_L \cdot INS\right),\sigma \left(1-\xi_\sigma \cdot INS\right)\right]$$

donde $L$, conocido como productividad latente, sigue la relación:

$$L = L + \gamma t$$

En esta expresión, $L$ depende del régimen de innovación, mientras que $\gamma$ captura el crecimiento de la productividad latente a lo largo del tiempo. Los parámetros $ \xi_ {L} $ y $ \xi_\sigma $ permiten incorporar al modelo la visión propuesta por Rodrik (2000), según la cual las instituciones desempeñan un papel crucial en la reducción de la incertidumbre y la variabilidad en la economía. En este contexto, $ \xi_ {L} $ refleja el aumento porcentual en la media de la distribución de la variable latente debido a la presencia de instituciones, mientras que $ \xi_\sigma $ captura la disminución porcentual en la varianza de dicha distribución. Estas variables destacan cómo un entorno institucional sólido puede fomentar una mayor estabilidad y previsibilidad en los procesos económicos.

Finalmente, según los resultados de cada investigación, la firma determina el nivel de tecnología del siguiente periodo:

$$A_{i(t+1)} = \max\{A_{it}, d_{imt} \cdot A_{it}^*, d_{int}\cdot \tilde A_{it}\}$$

Donde $A_{it}^* \equiv \max_i \left\{A_{i,(t-\tau)}\right\}$ es la tecnología máxima disponible en $\tau$ periodos atrás, donde $\tau$ es la duración de la patente. Además, $ \tilde A_{it}$ es la tecnología desarrollada en el proceso de innovación. 

Observemos que si:


$$\max\{A_{it}, d_{imt} \cdot A_{it}^*, d_{int}\cdot \tilde A_{it}\} = \tilde A_{it}$$


Esto implica que la firma ha "innovado" y patentado su producto, ya que ha superado las alternativas tecnológicas disponibles. En cualquier otro caso, no se genera una nueva patente.

La firma además de dedicar  recursos en investigación y desarrollo, decide aumentar su capital para el siguiente futuro siguiendo la fórmula:

$$K_{i(t+1)} = \left(1 + I_{it} -\delta\right)K_{it}$$

Donde $I_{it}$ representa la inversión bruta en capital por unidad de capital, y $\delta$ representa la depreciación del capital. En el momento $t = 0$, todas las firmas iniciales comienzan con el mismo capital $K_0$.
La formulación de la inversión está determinada por la variación del "markup", es decir, la diferencia entre el precio y el costo asociado con la venta de una unidad adicional en el mercado. Este cálculo considera no solo la elasticidad de la demanda, que refleja la reacción de los consumidores ante cambios en el precio, sino también la elasticidad de la oferta, que captura la respuesta de los demás productores ante estos mismos cambios.

El markup se define como:

$$\mu = \left\{\begin{align}\frac{\eta^d + (1-s)\eta^s}{\eta^d + (1-s)\eta^s - s} & & s<1\\
\\
\frac{p_t - c}{p_t} & & s = 1\end{align}\right.$$

Donde $\mu$ es el markup, $\eta^d$ y $\eta^s$ son las elasticidades de demanda y oferta respectivamente, y $s$ representa la participación de mercado de la firma.Este enfoque se inspira en la teoría marginalista, que postula que una firma continuará invirtiendo en capital adicional siempre que ello incremente sus beneficios netos.

De esta forma, la inversión bruta deseada por la empresa queda determinada por:

$$I^D = \delta + 1 - \mu \frac{c}{P_t\cdot A_{i(t+1)}}$$

Sin embargo, la inversión efectiva de la empresa está limitada por sus posibilidades financieras, que se definen mediante la siguiente función:

$$I^F = \left\{\begin{align} \delta + \pi_{it} & & \pi_{it} < 0\\
\\
\delta + (1 + b)\pi_{it} & & \pi_{it}\geq 0\end{align}\right.$$

Donde $b$ representa la proporción adicional de capital que los vancos están dispuestos a financiar cuando la firma tiene beneficios positivos. 

Cuando los beneficios son negativos $(\pi_{it} <0)$, la empresa se desfinancia, es decir, reduce su stock de capital. En cambio, cuando los beneficios son no negativos $(\pi_{it}\geq0)$, la firma no solo puede reinvertir su capital, sino también aumentar su capacidad de inversión en proporción a $b$b, gracias al financiamiento proporcionado por el sistema bancario.

Finalmente, la inversión efectiva de la firma esta determinada por:

$$I_{it} = \min\{I^D,I^F\}$$

Esto permite capturar cómo las decisiones de inversión reflejan las condiciones económicas y financieras específicas que enfrenta cada empresa, destacando las discrepancias entre la inversión deseada y la inversión factible, evitando las sobre simplificaciones del paradigma neoclásico.

A diferencia del modelo de Nelson y Winter (1982), Winter asume que las políticas de investigación se adaptan en función de los resultados obtenidos. En este marco, si el desempeño de una firma es inferior al beneficio promedio de las firmas en la industria $(\bar \pi_t)$, esta ajusta su política acercándose a la política promedio del resto de las firmas $(\bar r_{xt})$:

$$r_{ix(t+1)} =\begin{cases}r_{ixt} & X_{it}\geq \bar \pi_t\\ \\ r_{ixt} & X_{it}< \bar \pi_t & d_r = 0 \\ \\  1 - \beta r_{ixt} + \beta \bar r_{xt} + u_{ixt} & X_{it}< \bar \pi_t & d_r = 1\end{cases}\qquad \text{para } x \in [n, m]$$

El parámetro $\beta$ controla la velocidad o intensidad de adaptación hacia la política promedio.
El proceso de ajuste depende de una variable aleatoria $d_r$, que indica si la firma logra o no adaptar su política, con una probabilidad de éxito del 50% $(P(d_r= 1) = 0.5)$. Además, las firmas experimentan un error en el proceso de ajuste, modelado como una variable aleatoria con las siguientes distribuciones:

$$u_{int} \sim N(0,\sigma_n)\qquad\qquad u_{imt} \sim N(0,\sigma_m)$$

La entrada a la industria también sigue un proceso estocástico. En cada período, la entrada de firmas innovadoras e imitadoras se distribuye según un proceso de Poisson, independiente del tiempo. Las medias de estas distribuciones están dadas por:

$$M_t  = a_m\cdot E_m$$

$$N_t  = a_n\cdot E_n$$

donde $M_t$ y $N_t$ es  la  media de la distribución de imitación y de innovación respectivamente. Los parámetros $E_m$ y $E_n$ son considerados los esfuerzos que las firmas fuera de la industria realizan en investigación e innovación de esa industria.
Una vez determinada la cantidad de firmas que evalúan su ingreso, la decisión de entrar depende del costo de oportunidad asociado. Esto se evalúa según la siguiente condición:

$$P_t\cdot A_e - c> r_e+u_{et}$$

Donde $r_e$ representa el costo de oportunidad de la empresa entrante $A_e$ representa la cantidad de tecnología con la que entra y $u_{et}$ es un error de observación, con media cero y varianza $\sigma_u$. Esta formulación refleja que las firmas evalúan el costo esperado de ingresar a la industria frente a los beneficios proyectados, teniendo en cuenta la incertidumbre inherente al proceso, alineada con (Schmookler, 1954).

La tecnología con la que una firma entra depende de su naturaleza. Para las firmas innovadoras, la tecnología inicial proviene de la distribución de productividad latente descrita previamente, y su ingreso al mercado genera una patente. Mientras que para las firmas imitadoras, la tecnología inicial corresponde a la mejor tecnología disponible para imitar.

Por último, el capital inicial con el que una firma entra a la industria se determina a partir de una distribución con media $K^*$ y desviación estándar $\sigma_K$. De esta forma,  las firmas que entran a la industria presentan condiciones heterogéneas en cuento a su capital y tecnología.

Cuando el capital en el siguiente período no alcanza su nivel mínimo, $ K_{i,(t+1)} < K_{\min} $, o cuando el desempeño de las firmas resulta inferior al umbral establecido, $ X_{it} < X_{\min} $, dichas firmas no logran sostener su posición en el mercado y, como resultado, abandonan la industria. Este proceso refleja la incapacidad de las firmas para mantener la competitividad o generar recursos suficientes para operar eficientemente. La salida de la industria se formaliza mediante la condición $ K_{i,(t+\tau_K)} = 0 $, donde $ \tau_K \in (1, \infty) $, indicando un lapso de tiempo durante el cual el capital se agota completamente.

El modelo incorpora tanto la entrada como la salida de empresas. La entrada se modela como un proceso estocástico, en el cual las firmas evalúan la conveniencia de ingresar a la industria considerando los beneficios esperados y los costos de oportunidad, pero con cierto margen de error en su decisión. Por otro lado, la salida de las empresas está determinada por su desempeño y la disponibilidad de capital necesario para sostener sus actividades productivas.
